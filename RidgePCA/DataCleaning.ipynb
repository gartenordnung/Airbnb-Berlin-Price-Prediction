## Clean up NaN values

```python
nan_columns = df.columns[df.isna().any()].tolist()
print(nan_columns)
```

```python
# Check for existence of certain columns for debugging
if "has_tv" in df.columns:
    print("Column exists!")
else:
    print("Column does not exist.")

print(df["sqm"].isna().sum())
print(df["sqm"].notna().sum())
```

```python
df_no_nan = df.copy()

# If the number of listings of a host are missing, set them to 1.0 as an estimated default
df_no_nan["host_listings_count"] = df_no_nan["host_listings_count"].fillna(1.0)
df_no_nan["host_total_listings_count"] = df_no_nan["host_total_listings_count"].fillna(1.0)


# For bathrooms and bedrooms also assume 1.0, if value is missing
df_no_nan["bathrooms"] = df_no_nan["bathrooms"].fillna(1.0)
df_no_nan["bedrooms"] = df_no_nan["bedrooms"].fillna(1.0)


# For beds, assume one bed per bedrooms, if value is missing 
df_no_nan["beds"] = df_no_nan["beds"].fillna(df_no_nan["bedrooms"])


# For missing review scores, assume the average score of all other listings
review_columns = [
    "review_scores_rating",
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value"
]

for column in review_columns:
    mean_val = df_no_nan[column].mean()
    df_no_nan[column] = df_no_nan[column].fillna(mean_val)


# For reviews per month, tv size, wifi speed, sqm and the three numeric rates, also insert the means into the missing values
averageable_columns = [
    "reviews_per_month",
    #"sqm", # better fix written below
    "host_response_time_numeric",
    "host_response_rate_numeric",
    "host_acceptance_rate_numeric"
]

for column in averageable_columns:
    mean_val = df_no_nan[column].mean()
    df_no_nan[column] = df_no_nan[column].fillna(mean_val)
```

```python
# For wifi speed and tv size, set nan to average, if there is wifi/ tv, else set to 0.0
mean_wifi_speed_mbps = df_no_nan.loc[df_no_nan["has_wifi"] == 1.0, "wifi_speed_mbps"].mean()
mean_tv_size_inch = df_no_nan.loc[df_no_nan["has_tv"] == 1.0, "tv_size_inch"].mean()

# Fill NaNs where has_wifi == 0.0 with 0
df_no_nan.loc[(df_no_nan["has_wifi"] == 0.0) & (df_no_nan["wifi_speed_mbps"].isna()), "wifi_speed_mbps"] = 0.0
df_no_nan.loc[(df_no_nan["has_tv"] == 0.0) & (df_no_nan["tv_size_inch"].isna()), "tv_size_inch"] = 0.0

# Fill NaNs where has_wifi == 1.0 with the mean WiFi speed
df_no_nan.loc[(df_no_nan["has_wifi"] == 1.0) & (df_no_nan["wifi_speed_mbps"].isna()), "wifi_speed_mbps"] = mean_wifi_speed_mbps
df_no_nan.loc[(df_no_nan["has_tv"] == 1.0) & (df_no_nan["tv_size_inch"].isna()), "tv_size_inch"] = mean_tv_size_inch
```

```python
mean_val_pre = df_no_nan["sqm"].mean()
print(mean_val_pre)

# For sqm (square meters) fill nan values with average of the respective property type
property_type_columns = [col for col in df_no_nan.columns if col.startswith("property_type_")]

#print(len(property_type_columns))
#print(property_type_columns)

# Function to fill missing sqm values
def fill_sqm(row):
    if not np.isnan(row["sqm"]):
        return row["sqm"]  # if not nan, then keep value
    
    # Find property type and take mean over that group
    for col in property_type_columns:
        if row[col] == 1.0:
            # compute mean sqm for this property type
            mean_sqm = df_no_nan.loc[df_no_nan[col] == 1.0, "sqm"].mean()
            return mean_sqm
    
    # If no property type found, keep value as nan
    return row["sqm"]

# Apply the function over the rows
df_no_nan["sqm"] = df_no_nan.apply(fill_sqm, axis=1)
```

```python
print(df_no_nan["sqm"].isna().sum())
print(df_no_nan["sqm"].notna().sum())
```

180 listings do not have a property type listed -> for these, replace nan with the overall average

```python
mean_val = df_no_nan["sqm"].mean()
print(mean_val)
df_no_nan["sqm"] = df_no_nan["sqm"].fillna(mean_val)
```
